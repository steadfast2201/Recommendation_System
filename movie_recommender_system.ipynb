```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l         
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if(counter==3):
            break
        l.append(i['name'])
        counter+=1
    return l  
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(3)
```

```python
def fetch_direc(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_direc)
```

```python
movies.head(3)
```

```python
movies.rename(columns = {'crew':'Director'}, inplace = True)
```

# Tags formation Completion and Model training starting

```python
# now applying steming to remove similar words like (loved,love)
# specify same meaning
```

```python
pip install nltk
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['Tags'] = new_df['Tags'].apply(stem)
```

```python
# Will use CountVecetorizer...
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['Tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
# we will use cosine distance to calculate the differentiality of two
# vectors and we will not use Euclidean Distance because it fails 
# in higher Dimension.
```

```python
# To calculate Cosine distance we will use cosine Similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simi = cosine_similarity(vectors)
```

```python
sorted(simi[0],reverse=True)
```

```python
cosine_similarity(vectors).shape
```

```python
simi[1]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = simi[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman')
```
